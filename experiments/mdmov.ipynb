```python
import py3Dmol
from glob import glob


def view_trajectory(filename):
    with open(filename, "r") as f:
        data = f.read()
    view = py3Dmol.view(width=800, height=600)
    view.addModelsAsFrames(data, "xyz")
    view.setStyle({"stick": {}})
    view.zoomTo()
    view.animate({"loop": "backAndForth"})
    view.spin("y", 0.25)
    view.show()
    return view
```

```python
for file in glob("*.xyz"):
    print(file)
    view_trajectory(file)
```
