```python
import torch
import wandb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
id = "sw-apps/new_project/runs/1z1kitib"
api = wandb.Api()
df = api.run(id).history()
df
```

```python
loss = df["train/min_loss"]
loss = loss[~loss.isnull()].reset_index(drop=True)
loss
```

```python
loss.plot()
plt.xlabel("epoch")
plt.ylabel("MSE loss")
```
